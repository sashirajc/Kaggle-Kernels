## Homicide Reports 1980-2014

** Dataset Source: https://www.kaggle.com/murderaccountability/homicide-reports **

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('../data/database.csv')
```

```python
data
```

```python
data.info()
```

```python
data = data.rename(columns={'Record ID':'record','Agency Type':'agencyType','Crime Type':'crimeType',
                            'Crime Solved':'crimeSolved','Victim Sex':'vSex','Victim Age':'vAge','Victim Race':'vRace',
                            'Perpetrator Sex':'pSex','Perpetrator Age':'pAge','Perpetrator Race':'pRace',
                            'Perpetrator Ethnicity':'pEthnicity','Record Source':'rSource'
                           })
```

**Dangerous cities**

```python
stateData = data['State'].value_counts()
stateData[:6].plot(kind='bar')
```

California, Texas and New York have the most crime. This is not surprising since these states also have very high populations.

** Homicide by year **

```python
yearData = data['Year'].value_counts()
yearData.sort_index().plot(kind='line')
```

Homicide hit a peak in the first half of the 90s and reached an all time low by 2000. After a brief increase in the first half of 2000s homicides have been decreasing after 2006.
