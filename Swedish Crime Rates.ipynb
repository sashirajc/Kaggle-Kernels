## Swedish Crime Rates

** Dataset Source: Kaggle **

This dataset has a record of all crime in sweden between 1950-2015

```python
import pandas as pd
import matplotlib.pyplot as plt
```

To begin our data exploration, we will first load the data into a dataframe.

```python
data = pd.read_csv('reported.csv')
```

```python
data
```

We will plot the number of crimes as well as the population over the years to get an overall look at the data

```python
plt.plot(data['Year'],data['crimes.total'])
plt.xlabel('Year')
plt.ylabel('Number of crimes')
plt.show()

plt.plot(data['Year'],data['population'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

The crime rate has steadily increased over the years with the increase in population.

We will now explore each of the crime types in detail. For the sake of simplicity the crimes are broadly categorised into Assault/Murder, Sexual Crimes, Theft/Robbery, Fraud, Narcotics, Alcohol, Other Crimes

** Crimes of sexual nature **

```python
plt.plot(data['Year'],data['rape'])
plt.plot(data['Year'],data['sexual.offenses'])
plt.plot(data['Year'],data['crimes.total'])
plt.xlabel('Year')
plt.ylabel('Number of crimes')
plt.show()

plt.plot(data['Year'],data['rape'])
plt.plot(data['Year'],data['sexual.offenses'])
plt.xlabel('Year')
plt.ylabel('Number of crimes')
plt.show()
```

Crimes of sexual natures are only a small (almost insignificant portion of all the crimes that take place).

** Thefts/ Robberies **

Let us now explore the different types of theft/robberies.

```python
plt.plot(data['Year'],data['robbery'])
plt.plot(data['Year'],data['burglary'])
plt.plot(data['Year'],data['vehicle.theft'])
plt.plot(data['Year'],data['house.theft'])
plt.plot(data['Year'],data['shop.theft'])
plt.plot(data['Year'],data['out.of.vehicle.theft'])
plt.xlabel('Year')
plt.ylabel('Number of thefts/robberies')
plt.legend()
plt.show()

plt.plot(data['Year'],data['robbery'])
plt.plot(data['Year'],data['burglary'])
plt.plot(data['Year'],data['vehicle.theft'])
plt.plot(data['Year'],data['house.theft'])
plt.plot(data['Year'],data['shop.theft'])
plt.plot(data['Year'],data['out.of.vehicle.theft'])
plt.plot(data['Year'],data['crimes.total'])
plt.xlabel('Year')
plt.ylabel('Number of thefts/robberies')
plt.legend()
plt.show()
```

Thefts/Robberies also seem to be a very small portion of the total crimes that happen in Sweden. This suggests that most crimes are not of sexual nature and not robberies as well.

** Substance Abuse **

Let us now explore alcohol/drug related crimes

```python
plt.plot(data['Year'],data['narcotics'])
plt.plot(data['Year'],data['drunk.driving'])
plt.xlabel('Year')
plt.ylabel('Number of crimes')
plt.legend()
plt.show()

plt.plot(data['Year'],data['narcotics'])
plt.plot(data['Year'],data['drunk.driving'])
plt.plot(data['Year'],data['crimes.total'])
plt.xlabel('Year')
plt.ylabel('Number of crimes')
plt.legend()
plt.show()
```

Looks like drunk driving cases have not increased over the years but substance abuse has skyrocketed in the last decade with a short surge in the early 80s.

Substance abuse related crimes are still low compared to the total number of crimes

This could be a potential flaw in our analysis. This is because of our assumption that we might find one particular crime which is higher than all the others. It is completely possible that no distinct crime might emerge as the most common. Hence we should create new aggregate data for related crimes and visualize it to get a better idea of the type of crimes in Sweden.

** Aggregate data **

```python
data['Sexual Crimes'] = data['rape'] + data['sexual.offenses']
data['Theft'] = data['robbery'] + data['burglary'] + data['vehicle.theft'] + data['house.theft'] + data['shop.theft'] + data['out.of.vehicle.theft']
data['Personal Crimes'] = data['crimes.person'] + data['assault'] + data['murder']
data['Penal Code'] = data['crimes.penal.code'] + data['other.penal.crimes']
data['Substance Abuse'] = data['drunk.driving'] + data['narcotics']
```

```python
plt.plot(data['Year'],data['Sexual Crimes'])
plt.plot(data['Year'],data['Theft'])
plt.plot(data['Year'],data['Personal Crimes'])
plt.plot(data['Year'],data['Penal Code'])
plt.plot(data['Year'],data['Substance Abuse'])
plt.plot(data['Year'],data['fraud'])
plt.plot(data['Year'],data['criminal.damage'])
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.legend()
plt.show()
```

This gives us a clear picture of the type of crimes that are prevalent in Sweden and a vast majorty of them seem to be crimes against the penal code. It was quite difficult to find a proper definition for what really constitutes a crime against penal code but with a little background reading it seems to be crimes against the penal code which can constitute trivial crimes like disorderly behaviour in public.
